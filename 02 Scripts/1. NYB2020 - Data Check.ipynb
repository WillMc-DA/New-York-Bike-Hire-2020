## 6. Splitting out the date and time into a useable format

```python
#Creating a copy of the startime column, as Python sometimes messes up the original column that you are splitting from.
NYBmerge['StartTimeCOPY'] = NYBmerge['starttime']
```

```python
NYBmerge.info()
```

```python
#Converting to pandas timestamp
NYBmerge["StartTimeCOPY"] = pd.to_datetime(NYBmerge.starttime, dayfirst=True)
```

```python
#Spliting into the time and date columns
NYBmerge["start_date"] = NYBmerge["StartTimeCOPY"].dt.date
NYBmerge["Start_time"] = NYBmerge["StartTimeCOPY"].dt.time
```

```python
%%time
#Spliting the new Time column into Hours(start_hour), minutes(Sm) and seconds(Ss) columns
#The Seconds(Ss) column will have the milliseconds in it.
NYBmerge[['start_hour','Sm','Ss']] = NYBmerge['Start_time'].astype(str).str.split(':', expand=True).astype(str)
```

```python
NYBmerge.info()
```

```python
%%time
#Removing the new Minutes, Seconds, start_time and starttimecopy columns as they are now superfluous 
NYB2020 = NYBmerge.drop(columns = ['Sm', 'Ss', 'Start_time', 'StartTimeCOPY'])
```

```python
NYB2020.info()
```
